```python
import numpy as np
from sklearn.mixture import GaussianMixture
from scipy.stats import multivariate_normal
```

```python
def get_miniRR(row, means, sigmas, alpha=10, recurrence_rate=None, distance_vecindad=None, order=2):
    if recurrence_rate is None:
        recurrence_rate = [0]
    if distance_vecindad is None:
        distance_vecindad = []

    # Calcular distancia al centroide promedio
    dist = distance.euclidean(row, means)
    vecindad = sigmas * alpha

    # Evaluar si está dentro de la vecindad
    in_neighborhood = 1 if dist < vecindad else 0
    distance_vecindad.append(in_neighborhood)

    # Evaluar recurrencia según historial
    for i in range(order):
        if len(distance_vecindad) > i and distance_vecindad[-(i + 1)] == 1 and distance_vecindad[-i] == 1:
            recurrence_rate.append(recurrence_rate[-1] + 1)
            break
        elif i == order - 1:
            recurrence_rate.append(0)

    return recurrence_rate, distance_vecindad
```

# EM

```python
promedios = []
desves = []

initial_data = data[:, :10].T 
gmm = GaussianMixture(n_components=2, covariance_type='full', random_state=42)
gmm.fit(initial_data)


mus = gmm.means_

#mus[0] = np.zeros(mus[0].shape) + 0.13

sigmas = gmm.covariances_
pis = gmm.weights_

alpha = 0.05  # Learning rate

recurrence_rate = [0]
distance_vecindad = []
recurrence_rate_2 = [0]
distance_vecindad_2 = []

order = 3
# Online EM loop
i = 0
for x_t in data[:, :].T:
    # Step 2: E-step
    gamma = np.array([
        pis[k] * multivariate_normal.pdf(x_t, mean=mus[k], cov=sigmas[k])
        for k in range(2)
    ])
    
    gamma /= gamma.sum()  # Normalize

    # Step 3: M-step

    for k in range(2):
        diff = x_t - mus[k]
        mus[k] += alpha * gamma[k] * diff
        sigmas[k] += alpha * gamma[k] * (np.outer(diff, diff) - sigmas[k])
        pis[k] += alpha * (gamma[k] - pis[k])
        
    promedios.append(mus.copy())
    recurrence_rate, distance_vecindad = get_miniRR(x_t, mus[0], std_centros.mean(), 12, recurrence_rate, distance_vecindad, order)
    recurrence_rate_2, distance_vecindad_2 = get_miniRR(x_t, mus[1], std_centros.mean(), 12, recurrence_rate_2, distance_vecindad_2, order)
    

recurrence_rate = np.array(recurrence_rate)
distance_vecindad = np.array(distance_vecindad)
distance_vecindad_2 = np.array(distance_vecindad_2)
recurrence_rate_2 = np.array(recurrence_rate_2)
promedios = np.array(promedios)
```

# K- means

```python
initial_data = data[:, :3].T
k = 2  # número de clusters
centroids = initial_data[np.random.choice(initial_data.shape[0], k, replace=False)]

# Historial para graficar cómo se mueven
centroid_history = []

recurrence_rate = [0]
recurrence_rate_2 = [0]
distance_vecindad = []
distance_vecindad_2 =  []
order = 3
# Tasa de aprendizaje para actualizar los centroides
alpha = 0.01

# Loop online
for x_t in data.T:  # uno por uno
    # Paso 1: calcular distancia a cada centroide
    dists = [np.linalg.norm(x_t - c) for c in centroids]
    
    # Paso 2: asignar al más cercano
    closest = np.argmin(dists)
    
    # Paso 3: actualizar ese centroide
    centroids[closest] += alpha * (x_t - centroids[closest])
    
    # Guardar los centroides actuales
    centroid_history.append(centroids.copy())

    recurrence_rate, distance_vecindad = get_miniRR(x_t, centroids[0], 0.04, 12, recurrence_rate, distance_vecindad, order)
    recurrence_rate_2, distance_vecindad_2 = get_miniRR(x_t, centroids[1], 0.04, 12, recurrence_rate_2, distance_vecindad_2, order)
    

recurrence_rate = np.array(recurrence_rate)
distance_vecindad = np.array(distance_vecindad)
distance_vecindad_2 = np.array(distance_vecindad_2)
recurrence_rate_2 = np.array(recurrence_rate_2)
promedios = np.array(promedios)

centroid_history = np.array(centroid_history)

```

```python
import numpy as np
import matplotlib.pyplot as plt


n_neurons = 2              # Número de neuronas del SOM
input_dim = data.shape[0]  # Dimensión de entrada N de rasgos
alpha = 0.01               # Tasa de aprendizaje
sigma = 0.5                # Vecindad (se puede decrecer con el tiempo)

# Inicialización aleatoria de los pesos
weights = np.random.rand(n_neurons, input_dim)

# Guardamos historia para graficar
history_weights = []

# Función de vecindad gaussiana
def neighborhood(bmu_index, n_neurons, sigma):
    d = np.abs(np.arange(n_neurons) - bmu_index)
    return np.exp(-d**2 / (2 * sigma**2))

# Entrenamiento en línea
for x_t in data.T:
    # 1. Buscar la Best Matching Unit (BMU)
    dists = np.linalg.norm(weights - x_t, axis=1)
    bmu_idx = np.argmin(dists)

    # 2. Actualizar los pesos de todas las neuronas
    h = neighborhood(bmu_idx, n_neurons, sigma)
    for i in range(n_neurons):
        weights[i] += alpha * h[i] * (x_t - weights[i])
    
    history_weights.append(weights.copy())

# Convertimos historia a array
history_weights = np.array(history_weights)  # shape: (steps, n_neurons, input_dim)
```

```python
plt.figure(figsize = (8, 6))

for actividad in act:
    plt.scatter(data[0, actividad[0]: actividad[1]], data[1, actividad[0]: actividad[1]], color  = "red", alpha=0.1)

for reposo in rep:
     plt.scatter(data[0,reposo[0]: reposo[1]], data[1, reposo[0]: reposo[1]], color  = "blue", alpha=0.1)


for i in range(n_neurons):
    plt.plot(history_weights[:, i, 0], history_weights[:, i, 1], label=f'Neuron {i}', alpha = 0.4)
    plt.scatter(history_weights[-1, i, 0], history_weights[-1, i, 1], marker='*', s=200)

plt.title("Evolución de las neuronas del SOM")
plt.legend()
plt.grid(False)
plt.show()
```
